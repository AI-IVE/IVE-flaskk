```python
import pandas as pd
import time
import utils 
# CSV 파일 읽기 1차 한->일 했을 때 
df = pd.read_csv('./wtech_data/topic_content.csv' ,encoding='cp949')

# df = pd.read_csv('./translated_data.csv' )
# # 번역 함수 정의 (실제 API 호출 함수로 대체해야 함)


system_prompt_translate = utils.load_prompt("187", lang="ja")
    
# 번역 수행 한국어-> 일본어
df['gpt_trans'] = df['topic_content'].apply(lambda x: utils.api_openai_native(system_prompt_translate, x, "gpt-4o") if pd.notna(x) else '')
df['claude_trans'] = df['topic_content'].apply(lambda x: utils.api_claude_native(system_prompt_translate, x, "claude-3.5-sonnet") if pd.notna(x) else '')

# # 번역 수행 일본어 -> 한국어 
# df['gpt_trans_jp_ko'] = df['gpt_trans'].apply(lambda x: utils.api_openai_native(system_prompt_translate, x, "gpt-4o") if pd.notna(x) else '')
# df['claude_trans_jp_ko'] = df['claude_trans'].apply(lambda x: utils.api_claude_native(system_prompt_translate, x, "claude-3.5-sonnet") if pd.notna(x) else '')


# 결과 확인
print(df[['topic_content', 'gpt_trans', 'claude_trans']])

# 결과를 CSV 파일로 저장
df.to_csv('translated_data.csv', index=False)
```
